```python
import pandas as pd
```

```python
df1 = pd.read_excel('./datasets/Aracaju.xlsx')
df2 = pd.read_excel('./datasets/Fortaleza.xlsx')
df3 = pd.read_excel('./datasets/Natal.xlsx')
df4 = pd.read_excel('./datasets/Recife.xlsx')
df5 = pd.read_excel('./datasets/Salvador.xlsx')
```

```python
df = pd.concat([df1, df2, df3, df4, df5])
df.head()
```

```python
#Convertendo a coluna de data em data 

df["Data"] = pd.to_datetime(df["Data"])
```

```python
#Criando a coluna receita novamente

df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
#Agrupando os dados por ano

df.groupby(df["Data"].dt.year)["Receita"].sum()
df.sample(5)
```

```python
#Criando  uma nova coluna com o ano

df["Ano_Venda"] = df["Data"].dt.year
```

```python
#Extraindo o mês e o dia 

df["mês_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
df.sample(5)
```

```python
#Calculando a diferença em dias entre um venda e a venda minima

df["Diferença_dias"] = df["Data"] - df["Data"].min()
df.sample(5)
```

```python
#Criando a coluna trimestre 

df["Trimestre_venda"] = df["Data"].dt.quarter
df.sample(5)
```

```python
#Filtrando as vendas de 2019 do mês de março

vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
vendas_marco_19.sample(5)
```
